Concatenación y división de arrays

Todas las rutinas anteriores funcionaron en arrays individuales. También es posible combinar varios arrays en uno solo o, por el contrario, dividir un array en varios. La concatenación, o unión de dos arrays en NumPy, se realiza principalmente a través de la rutina <code>np.concatenate</code>


```python
import numpy as np
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
z = np.concatenate([x, y])
print(z)
```


También se puede utilizar para arrays bidimensionales

```python
grilla = np.array([[1, 2, 3], [4, 5, 6]])

z1 = np.concatenate([grilla,grilla])            # concatenar a lo largo del primer eje
print(z1)
print(5*'-')
z2 = np.concatenate([grilla,grilla], axis=1)    # concatenar a lo largo del segundo eje
print(z2)
```

# Matplotlib

Matplotlib es probablemente el paquete de Python más utilizado para gráficos 2D. Proporciona una forma rápida de visualizar datos de Python y cifras con calidad de publicación en muchos formatos. Esta es una colección de funciones de estilo de comando que hacen que Matplotlib funcione como MATLAB. Cada función de pyplot realiza algún cambio en una figura: por ejemplo, crea una figura, crea un área de trazado en una figura, traza algunas líneas en un área de trazado, decora la trama con etiquetas, etc.

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Graficos Simples

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C)
plt.plot(X, S)
plt.xlim((3,-3))
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C)
plt.plot(X, S)

plt.show()
```

```python
plt.figure(figsize=(10, 6), dpi=80)
plt.plot(X, C, color="blue", linewidth=1, linestyle="-")
plt.plot(X, S, color="red",  linewidth=5, linestyle=":") # -- -. etc
plt.show()
```

```python
plt.plot(X, C)
plt.plot(X, S)
plt.xlim(X.min() * 5, X.max() * 4)
plt.ylim(C.min() * 1.1, C.max() * 1.1)
plt.show()
```

```python
plt.plot(X, C)
plt.plot(X, S)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']) #opcional
plt.yticks([-1, 0, +1])
plt.show()
```

```python
plt.plot(X, C)
plt.plot(X, S)
ax = plt.gca()
ax.spines['right'].set_color('blue') # Marco derecho
ax.spines['top'].set_color('none')  #Marco izquierda
ax.xaxis.set_ticks_position('bottom')                    # etiquete los numeros
ax.spines['bottom'].set_position(('data',0)) # el eje de abajo a la posicion cero
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.show()
```

```python
plt.plot(X, C)
plt.plot(X, S)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine") # uso de leyendas
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label="sine")
plt.legend(loc='upper left')  # donde poner la caja

plt.grid() # mostar cuadricula

plt.show()
```

```python
plt.plot(X, C)
plt.plot(X, S)

ax = plt.gca()
ax.spines['right'].set_color('none') # Marco derecho
ax.spines['top'].set_color('none')  #Marco izquierda
ax.xaxis.set_ticks_position('bottom')                    # etiquete los numeros
ax.spines['bottom'].set_position(('data',0)) # el eje de abajo a la posicion cero
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

# marcar un punto dentro de la grafica

t = 2 * np.pi / 3 # Punto a marcar

# para coseno
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle="--") # Grafica las lineas
plt.scatter([t, ], [np.cos(t), ], 50, color='blue')                           # Grafica el punto
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',                         # Se crea el texto
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16, # Grafica el texto
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")) # Grafica el vector

# Para seno
plt.plot([t, t],[0, np.sin(t)], color='red', linewidth=2.5, linestyle="--")
plt.scatter([t, ],[np.sin(t), ], 50, color='red')
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 -y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X,Y = np.meshgrid(x, y) # Matriz cordenada   help(np.meshgrid)

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap=plt.cm.hot)
C = plt.contour(X, Y, f(X, Y), 8, colors='black')
plt.clabel(C, inline=1, fontsize=10)


plt.xticks([])
plt.yticks([])
plt.show()
```
